```python
import numpy as np
import pandas as pd
```

```python
# x, y
x=np.random.randn(100, 1)
a=1000
y=x*a

a_hat=np.random.randint(1)

for _ in range(10000):
    y_hat=x*a_hat # 가설함수: 추측 y값
    err=np.mean((y-y_hat)**2)
    diff_err=np.mean(2*x*(y_hat-y))
    a_hat -= diff_err*1e-4 #경사하강
    print(f' a_hat: {a_hat} \n err: {err} \n→ 미분값: {diff_err}\n')
```

```python
#x 다시 정의, bias 있는 선형함수
x=np.random.randn(1000,1)
a=18
b=1800
y=x*a+b 

a_hat=np.random.randn(1)
b_hat=0

epochs = 10000
learning_rate=1e-3
for _ in range(epochs):
    #가설함수
    y_hat=x*a_hat+b_hat

    #err: 비용함수(cost)
    err=np.mean((y_hat - y)**2)

    #미분값
    diff_a_err = np.mean(x*(y_hat-y)) #혹은 2*x*(y_hat-y), 근데 상관없음
    diff_b_err=np.mean(2*(y_hat-y))

    #미분값을 적용해서 경사하강(Gradient Descent)
    a_hat-=diff_a_err*learning_rate
    b_hat-=diff_b_err*learning_rate
```

```python
a_hat, b_hat
```

```python
### 총 정리 ###
##초기 설정
# 초깃값
x=np.random.randn(1000,1)
w=100
b=150
y=x*w+b

#학습을 위한 초깃값
w_hat=np.random.randn(1)
b_hat=np.random.randn(1)

#cost function(error) & diff function
def err_w(w_hat):
    y_hat=x*w_hat+b_hat
    return np.mean((y-y_hat)**2)
def err_b(b_hat):
    y_hat=x*w_hat+b_hat
    return np.mean((y-y_hat)**2)
def diff(f, x):
    h=1e-4
    fx=f(x)
    fxh=f(x+h)
    return (fxh-fx)/h

##학습
epochs=10000
lr=1e-3
for _ in range(epochs):
    y_hat=x*w_hat+b_hat
    w_hat -= diff(err_w, w_hat)*lr
    b_hat -= diff(err_b, b_hat)*lr
    
w_hat, b_hat
```

```python
### 총 정리-cost fn 통일(w, b 둘 다 받기) ###
##초기 설정
# 초깃값
x=np.random.randn(1000,1)
w=100
b=150
y=x*w+b

#학습을 위한 초깃값
w_hat=np.random.randn(1)
b_hat=np.random.randn(1)

#cost function(error) & diff function
def err(w, b):
    y_hat=x*w+b
    return np.mean((y-y_hat)**2)
def diff(f, w, b):
    h=1e-4
    fx=f(w,b)
    fxh_w=f(w+h,b)
    fxh_b=f(w,b+h)
    dw=(fxh_w-fx)/h
    db=(fxh_b-fx)/h
    return dw, db
```
