#### 객체에 대한 이해가 부족한 것으로 보여 객체에 대한 추가 설명

```python
class Calc:
    
    def __init__(self,x,y):
        self.x = x
        self.y = y
        ### 객체 생성시 add함수도 자동으로 실행되게끔 하여 add_result가
        ### 만들어지게끔함 고로 클래스.add()와 클래스.add_result로 둘다 합을
        ### 불러올 수 있다.
        self.add()
    
    def add(self):
        self.add_result = self.x + self.y
        return self.add_result
    
    def __call__(self):
        return self.x*self.y
```

```python
Calc_class = Calc(3, 5)
### 속성 값
print(Calc_class.x)
print(Calc_class.y)
print(Calc_class.add_result)

### 클래스 내 함수(메소드)
print(Calc_class.add)
print(Calc_class.add())
```

```python
### __call__은 클래스의 이름()으로 불렀을 때 그 리턴값이 나옴
Calc_class()
```

#### 미분 직접해보기
- 미분값은 { f(x+h) - f(x) } / { (x+h) - (x) }
- = { f(x+h) - f(x) } / h
- 1e-4는 매우 작은 값

```python
# 수치미분
### 위를 나타내는 함수
def diff(f,x):
    h = 1e-4
    fxh = f(x+h)
    fx = f(x)
    return (fxh-fx)/h
```
