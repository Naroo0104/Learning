```python
# get_pw 함수 작성
import hashlib
pw = '1234'
pw_to_bytes = bytes(pw, 'utf8')
m = hashlib.sha256() #sha: secure hash algorythm
m.update( pw_to_bytes )
print( m.hexdigest() )
```

```python
try: #try block
    li=[]
    li[5]
#     4/0
    print('hello~')
except ZeroDivisionError as e:
    print('0으로 나누지 마세욧!',e)
except: #except block
    print('우웩~!')
```

예외처리 할 일

○try, except 하기 전에 미리 돌려보면서 오류 종류 확인해두자!
○try-except할 때 원인결과 둘 다 묶어줘야 함
○하면 안 될 코드: try-except에 pass 써버리면 오류 계속 발생

```python
import sqlite3
conn = sqlite3.connect('geoje.db')
cursor = conn.cursor()
sql = '''
SELECT id, name, desc, mapinfo, latitude, longitude
FORM tour
''' # FROM tour
cursor.execute(sql) #에러 원인 ##try-except할 때 원인결과 둘 다 묶어줘야 함
result = cursor.fetchmany(10) #fetchmany: data 가져와

for row in result:
    print(row[1], row[4], row[5])

cursor.close()
conn.close()
```

```python
import sqlite3
conn = sqlite3.connect('geoje.db')
cursor = conn.cursor()
sql = '''
SELECT id, name, desc, mapinfo, latitude, longitude
FORM tour
''' # FROM tour
try:
    cursor.execute(sql) #에러 원인 
    result = cursor.fetchmany(10) #fetchmany: data 가져와

    for row in result:
        print(row[1], row[4], row[5])
except:
    print('오류 발생!')
cursor.close()
conn.close()
```
