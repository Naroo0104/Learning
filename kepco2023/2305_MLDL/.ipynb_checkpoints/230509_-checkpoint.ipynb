```python
# 전략 무
def monti():
    windows=np.arange(3) #range는 제너레이터, arange는 값
    win_window=np.random.choice(windows,1)
    user_window=np.random.choice(windows,1)
    
    if win_window == user_window:
        open_window=windows[windows != user_window][0] #아무거나 말고 그냥 첫번째거 가져와서 쓴다.
    else:
        open_window=windows[(windows != user_window) & (windows != win_window)]
    return win_window, user_window
```

```python
# 3전략
def monti(strategy):
    windows=np.arange(3) #range는 제너레이터, arange는 값
    win_window=np.random.choice(windows,1)
    user_window=np.random.choice(windows,1)
    
    if win_window == user_window:
        open_window=windows[windows != user_window][0] #아무거나 말고 그냥 첫번째거 가져와서 쓴다.
    else:
        open_window=windows[(windows != user_window) & (windows != win_window)]
        
    if strategy == 'change':
        choice_window = windows[(windows != user_window) & (windows != open_window)]
    elif strategy=='random':
        choice_window = np.random.choice(windows[windows != open_window],1)
    else:
        choice_window = user_window
    return choice_window == win_window
```

```python
def fibo(n):
    if n <= 1:
        return n
    return fibo(n-1) + fibo(n-2)
```

```python
from functools import cache
```

```python
@cache #데코레이터, 내가 만든 fibo 함수가 cache까지 갖고 감
def fibo(n):
    if n <= 1:
        return n
    return fibo(n-1) + fibo(n-2)
```

```python
fibo(400) #fibo가 cache가 돼서 금방 결과가 나옴
```

```python
from functools import lru_cache
```

```python
@lru_cache(maxsize=10)
def fibo(n):
    if n <= 1:
        return n
    return fibo(n-1) + fibo(n-2)
```

```python
def cost(w, b):
    y_hat=sigmoid(np.dot(x,w)+b)
    return np.sum(y*np.log(y_hat)+(1-y)*np.log(1-y_hat))

def diff(func, w, b):
    h=1e-4
    grad_a = np.zeros_like(w)
    for i in range(w.shape[0]):
        for j in range(w.shape[1]):
            tmp = w[i,j]
            fx=func(w,b)
            w[i,j]=tmp+h
            fxh=func(w,b)
            grad_a[i,j]=(fxh-fx)/h
            w[i,j]=tmp # x, y= y, x 하는 방법(copy 만들기)
    fx=func(w,b)
    fxh=func(w,b+h)
    grad_b=(fxh-fx)/h
    return grad_a, grad_b

def sigmoid(x):
    return 1/(1+np.exp(-x))

class doggy:
    def __init__(self):
        self.w=None
        self.b=None
        print('bowwow')
        
    def fit(self, x, y):
        self.x=x
        self.y=y
        self.b=0
        self.w=np.random.randn(self.x.shape[1], self.y.shape[1])
        lr=1e-3
        y_hat=sigmoid(np.dot(self.x, self.w)+self.b)##가설함수->예측
        
        for _ in range(10000):
            self.w-=diff(cost, self.w, self.b)[0]*lr
            self.b-=diff(cost, self.w, self.b)[1]*lr
        
        return f'doggy model complete'
    def predict(self,x):
        return np.where(sigmoid(np.dot(x, self.w)+self.b) > .5, 1, 0)
    
    def score(self, x, y):
        y_hat = np.where(sigmoid(np.dot(x, self.w)+self.b) > .5, 1, 0)
        return np.around(sum(y_hat == y)/y.size, 3)
    
```

```python
def camelcase(x):
    return ''.join([word.capitalize() for word in x.split('_')])
```

```python
def mapper(func):
    def wrapper(list_arg):
        return [func(word) for word in list_arg]
    return wrapper
```

```python
@wrapper
def camelcase(x):
    return ''.join([word.capitalize() for word in x.split('_')])
```
